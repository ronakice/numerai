```python

lQuestions = list(qap["question"])
print("Number of Questions: " , len(lQuestions))
```

```python
whWords = ["who", "what", "when","where","how","why","which"]
ctnWh = 0
print("Non wh.* or how questions or mispells:\n\n")
for i in lQuestions:
    if not any_in(whWords,i.lower()):
        print(i)
        ctnWh += 1
print("Non wh* or how question count: ", ctnWh)
```

```python
whCounts = defaultdict(int)
exWh = {}
for i in lQuestions:
    for j in i.lower().split():
        if j in whWords:
            whCounts[j]+=1
            if whCounts[j]==42 or whCounts[j]==1001:
                print(j,i)
            break
```
